# ðŸ§© Notebook 01: Setup, Styling, and First Visualizations in Seaborn

```python
# ðŸ“¦ Setup & Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path

# âœ… Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

# ðŸ§° Custom utilities
from plot_utils import (
    apply_theme,
    stylize_plot,
    add_reference_line,
    save_fig,
    annotate_points
)

# ðŸŽ¨ Apply project-wide theme
apply_theme(style="whitegrid", context="notebook", palette="deep")

print("âœ… Seaborn Project Utils Loaded")
```

```python
# Load and inspect built-in 'tips' dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Histogram with KDE
plt.figure(figsize=(8, 5))
sns.histplot(data=tips, x="total_bill", kde=True, bins=20)

stylize_plot(
    title="Distribution of Total Bill",
    xlabel="Total Bill ($)",
    ylabel="Frequency"
)

save_fig("../exports/01_setup/tips_total_bill_histogram.png")
plt.show()
```

```python
# Scatterplot with hue and style
plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", style="smoker", s=100)

stylize_plot(
    title="Total Bill vs Tip by Gender & Smoking Status",
    xlabel="Total Bill ($)",
    ylabel="Tip ($)"
)

save_fig("../exports/01_setup/tips_bill_vs_tip_scatter.png")
plt.show()
```

```python
# Load dummy sales dataset
sales = pd.read_csv("../datasets/sales_data.csv")
sales.head()
```

```python
# Boxplot of Total Sales by Region
plt.figure(figsize=(8, 5))
sns.boxplot(data=sales, x="Region", y="Total Sales")

stylize_plot(
    title="Total Sales Distribution by Region",
    xlabel="Region",
    ylabel="Sales ($)"
)

add_reference_line(value=20000, axis='h', color="red")
save_fig("../exports/01_setup/sales_by_region_boxplot.png")
plt.show()
```

```python
# Histogram of units sold
plt.figure(figsize=(8, 5))
sns.histplot(data=sales, x="Units Sold", hue="Region", multiple="stack", bins=20)

stylize_plot(
    title="Units Sold by Region",
    xlabel="Units Sold",
    ylabel="Count"
)

save_fig("../exports/01_setup/sales_units_sold_by_region.png")
plt.show()
```
