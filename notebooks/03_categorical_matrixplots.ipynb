# ðŸ§© Notebook 03: Categorical and Matrix Visualizations in Seaborn

```python
# ðŸ“¦ Setup & Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path

# âœ… Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

# ðŸ§° Custom utilities
from plot_utils import (
    apply_theme,
    stylize_plot,
    add_reference_line,
    save_fig,
    annotate_points
)

# ðŸŽ¨ Apply project-wide theme
apply_theme()

print("âœ… Seaborn Project Utils Loaded")
```

```python
# Load dummy employee data

employee = pd.read_csv("../datasets/employee_data.csv")
employee.head()
```

```python
# ðŸ“Š Countplot (Categorical frequency)

plt.figure(figsize=(7, 5))
sns.countplot(data=employee, x="Department", hue="Gender")

stylize_plot(
    title="Employee Count by Department & Gender",
    xlabel="Department",
    ylabel="Count"
)

save_fig("../exports/03_categorical/employee_department_gender_count.png")
plt.show()
```

```python
# ðŸ“Š Boxplot (Salary distribution by department)

plt.figure(figsize=(7, 5))
sns.boxplot(data=employee, x="Department", y="Salary", hue="Gender")

stylize_plot(
    title="Salary Distribution by Department & Gender",
    xlabel="Department",
    ylabel="Salary ($)",
    rotate_xticks=10
)

save_fig("../exports/03_categorical/employee_salary_boxplot.png")
plt.show()
```

```python
# ðŸ“Š Violinplot

plt.figure(figsize=(7, 5))
sns.violinplot(data=employee, x="Department", y="Salary", inner="quartile", hue="Gender", split=True)

stylize_plot(
    title="Salary Distribution (Violinplot)",
    xlabel="Department",
    ylabel="Salary ($)"
)

save_fig("../exports/03_categorical/employee_salary_violinplot.png")
plt.show()
```

```python
# ðŸ“Š Swarmplot

plt.figure(figsize=(8, 5))
sns.swarmplot(data=employee, x="Department", y="Experience (Years)", hue="Gender")

stylize_plot(
    title="Experience by Department (Swarmplot)",
    xlabel="Department",
    ylabel="Experience (Years)"
)

save_fig("../exports/03_categorical/employee_experience_swarmplot.png")
plt.show()
```

```python
# ðŸ“Š Stripplot

plt.figure(figsize=(8, 5))
sns.stripplot(data=employee, x="Department", y="Experience (Years)", jitter=True, hue="Gender")

stylize_plot(
    title="Experience by Department (Stripplot)",
    xlabel="Department",
    ylabel="Experience (Years)"
)

save_fig("../exports/03_categorical/employee_experience_stripplot.png")
plt.show()
```

```python
# ðŸ“Š Pointplot (Trend of Salary per Dept)

plt.figure(figsize=(8, 5))
sns.pointplot(data=employee, x="Department", y="Salary", hue="Gender", dodge=True, markers=["o", "s"])

stylize_plot(
    title="Avg Salary Trend by Department & Gender",
    xlabel="Department",
    ylabel="Salary ($)"
)

save_fig("../exports/03_categorical/employee_salary_pointplot.png")
plt.show()
```

```python
# ðŸ§ª Load Matrix Data: Correlation Matrix (from sales_data.csv)

sales = pd.read_csv("../datasets/sales_data.csv")
corr_matrix = sales[["Units Sold", "Unit Price", "Total Sales"]].corr()
corr_matrix
```

```python
# ðŸ“Š Heatmap

plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5)

stylize_plot(
    title="Correlation Heatmap: Sales Metrics",
    xlabel="",
    ylabel=""
)

save_fig("../exports/03_categorical/sales_correlation_heatmap.png")
plt.show()
```

```python
# ðŸ“Š Clustermap

clustergrid = sns.clustermap(
    corr_matrix,
    annot=True,
    cmap="viridis",
    linewidths=0.5
)

clustergrid.fig.suptitle("Sales Metrics Clustermap", y=1.05)

# Save using our utility
save_fig("../exports/03_categorical/sales_clustermap.png")
```
