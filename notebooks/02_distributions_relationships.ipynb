```python
# ðŸ“¦ Setup & Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
from pathlib import Path

# âœ… Setup path to use utils
PROJECT_ROOT = Path.cwd().parent
UTILS_DIR = PROJECT_ROOT / "utils"
if str(UTILS_DIR) not in sys.path:
    sys.path.insert(0, str(UTILS_DIR))

# ðŸ§° Custom utilities
from plot_utils import (
    apply_theme,
    stylize_plot,
    add_reference_line,
    save_fig,
    annotate_points
)

# ðŸŽ¨ Apply project-wide theme
apply_theme()

print("âœ… Seaborn Project Utils Loaded")
```

```python
# Load tips dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# ðŸ“Š Histogram + KDE

plt.figure(figsize=(8, 5))
sns.histplot(data=tips, x="total_bill", kde=True, bins=30)

stylize_plot(
    title="Histogram with KDE: Total Bill",
    xlabel="Total Bill ($)",
    ylabel="Density"
)

save_fig("../exports/02_distributions/dist_total_bill_kde.png")
plt.show()
```

```python
# ðŸ“Š ECDF Plot
plt.figure(figsize=(8, 5))
sns.ecdfplot(data=tips, x="total_bill", hue="sex")

stylize_plot(
    title="ECDF: Total Bill by Gender",
    xlabel="Total Bill ($)",
    ylabel="Proportion"
)

save_fig("../exports/02_distributions/ecdf_total_bill.png")
plt.show()
```

```python
# ðŸ“Š Rug Plot (Layered)

plt.figure(figsize=(8, 5))
sns.kdeplot(data=tips, x="total_bill", fill=True)
sns.rugplot(data=tips, x="total_bill", height=0.05)

stylize_plot(
    title="KDE with Rugplot: Total Bill",
    xlabel="Total Bill ($)",
    ylabel="Density"
)

save_fig("../exports/02_distributions/kde_rug_total_bill.png")
plt.show()
```

```python
# ðŸ“Š Scatterplot with hue, size, style

plt.figure(figsize=(8, 5))
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time", size="size", style="sex")

stylize_plot(
    title="Scatterplot: Tip vs Total Bill",
    xlabel="Total Bill ($)",
    ylabel="Tip ($)"
)

save_fig("../exports/02_distributions/scatter_tip_vs_bill_size_time.png")
plt.show()
```

```python
# ðŸ“ˆ Line Plot (Smoothed trend)

plt.figure(figsize=(8, 5))
sns.lineplot(data=tips, x="size", y="tip", ci="sd")

stylize_plot(
    title="Line Plot: Average Tip by Table Size",
    xlabel="Table Size",
    ylabel="Average Tip ($)"
)

save_fig("../exports/02_distributions/lineplot_tip_by_size.png")
plt.show()
```

```python
# ðŸ“ˆ Regression Plot + Residuals

plt.figure(figsize=(8, 5))
sns.regplot(data=tips, x="total_bill", y="tip", scatter=True, ci=95)

stylize_plot(
    title="Regression Plot: Tip vs Total Bill",
    xlabel="Total Bill ($)",
    ylabel="Tip ($)"
)

save_fig("../exports/02_distributions/regplot_tip_vs_bill.png")
plt.show()
```

```python
# Residual Plot
plt.figure(figsize=(8, 5))
sns.residplot(data=tips, x="total_bill", y="tip")

stylize_plot(
    title="Residual Plot: Tip vs Total Bill",
    xlabel="Total Bill ($)",
    ylabel="Residuals"
)

save_fig("../exports/02_distributions/residplot_tip_vs_bill.png")
plt.show()
```
